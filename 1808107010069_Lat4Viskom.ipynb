```python
#Deteksi wajah pada gambar
import cv2
from matplotlib import pyplot as plt

gambar = ["18","16","17","14","5","6","19","11","9","12"]
jumlah_wajah = [3, 9, 8, 5, 5, 5, 6, 5, 15, 4]

for image, wajah in zip(gambar, jumlah_wajah):
  #download folder cascade di https://drive.google.com/file/d/1AcMPiaExIdzCmBc2E24exnPtwIB3_ir6/view?usp=sharing
  # face_cascade = cv2.CascadeClassifier('./cascades/haarcascade_frontalface_default.xml')
  #atau
  face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades +'haarcascade_frontalface_default.xml')

  #download gambar di https://drive.google.com/file/d/1UceTLyCk38JRqOqq7mJZazi9ZdEQVu1U/view?usp=sharing
  img = cv2.imread('/content/'+image+'.jpg')
  gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  faces = face_cascade.detectMultiScale(gray, 1.08, 5)

  jumlahwajah = wajah

  wajah_terdeteksi = 0
  for (x, y, w, h) in faces:
      img = cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
      wajah_terdeteksi+=1
      
  # cv2.namedWindow('Woodcutters Detected!')
  plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)),plt.show()
  cv2.imwrite('/content/'+image+'-detected.jpg', img)
  cv2.waitKey(0)
  cv2.destroyAllWindows()

  #lalu hitung probabilitas setiap wajah pada suatu gambar yang berhasil terdeteksi (secara manual)
  Probabilitas = wajah_terdeteksi/jumlahwajah
  print("Probabilitas = ", wajah_terdeteksi,"/",jumlahwajah, '=', Probabilitas)
```
